```python
#libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as dick
```

```python
# Generate 10,000 random numbers between 0 and 100
data = np.random.normal(50, 20, 10000)
```

```python
# first quartil
q1 = np.percentile(data, 25)

# second quartil
q2 = np.percentile(data, 50)

# third quartil
q3 = np.percentile(data, 75)

print("1. Quartil: ", q1)
print("Half: ", q2)
print("3. Quartil: ", q3)
```

```python
# Plot histogram with 50 bins and green color
plt.figure(figsize=(6,6))
plt.hist(data, bins=100, color='pink')

# Add title and labels
plt.title("Histogram of 10,000 randomly generated numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(linestyle='-', linewidth=0.1)

# Show the plot
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(7,1.4))
plt.ticklabel_format(style='plain')
dick.boxplot(x=data)

plt.title("Boxplot of randomly generated numbers")
plt.xlabel("Value")
plt.show()
```
